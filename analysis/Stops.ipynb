## Récupération des données des arrêts de bus et de tram

```python
from botigo import lacub

client = lacub.LacubAPI()
stops = client.get_stops()
```

```python
import pandas as pd
stops = pd.DataFrame(stops)
stops.head()
```

```python
stops.type.unique()
```

```python
tram_stops = stops[stops.type == 'TRAM']
tram_stops.head()
```

```python
import folium

BORDEAUX_LOC = {
    'lat': 44.8404400,
    'lon': -0.5805000
}

map_bdx = folium.Map(location=[BORDEAUX_LOC['lat'], BORDEAUX_LOC['lon']], zoom_start=12)

for row in tram_stops.itertuples():
    folium.Marker(location=[row.latitude, row.longitude], popup=row.libelle).add_to(map_bdx)

map_bdx
```

Il serait intéressant d'affecter une couleur à chaque ligne de tramway. Il nous faut donc associer chaque arrêt à son numéro de ligne.

```python
lines_paths = pd.DataFrame(client.get_lines_paths())
lines_paths = lines_paths[lines_paths.type == 'TRAM']
lines_paths
```

```python
lines_paths.rs_sv_ligne_a.unique()
```

On peut ensuite effectuer la jointure sur le numéro de ligne `rs_sv_ligne_a` pour avoir le nom de la ligne.

```python
lines = client.get_lines()
lines = pd.DataFrame(lines)
tram_lines = lines[lines.type == 'TRAM']
tram_lines.head()
```

```python
lines_paths = lines_paths.merge(tram_lines[['active', 'gid', 'libelle']], left_on='rs_sv_ligne_a', right_on='gid')
lines_paths.head(10)
```

```python
path_stops = pd.DataFrame(client.get_path_stops(gid=267436931))
path_stops = pd.DataFrame(path_stops)
path_stops.head()
```

Nous pouvons désormais joindre chaque arrêt avec les chemins de lignes et les arrêts sur les chemins.

```python
df = tram_stops.merge(path_stops, left_on='gid', right_on='gid', suffixes=('_ts', '_ps'))
df.head()
```
