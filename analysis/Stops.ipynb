## Récupération des données des arrêts de bus et de tram

```python
from botigo import lacub

client = lacub.LacubAPI()
stops = client.get_stops()
```

```python
import pandas as pd
stops = pd.DataFrame(stops)
stops.head()
```

```python
stops.type.unique()
```

```python
tram_stops = stops[stops.type == 'TRAM']
tram_stops.head(15)
```

```python
import folium

BORDEAUX_LOC = {
    'lat': 44.8404400,
    'lon': -0.5805000
}

map_bdx = folium.Map(location=[BORDEAUX_LOC['lat'], BORDEAUX_LOC['lon']], zoom_start=12)

for row in tram_stops.itertuples():
    folium.Marker(location=[row.latitude, row.longitude], popup=row.libelle).add_to(map_bdx)

map_bdx
```
